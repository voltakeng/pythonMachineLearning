```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn import metrics
```

```python
dataset = pd.read_csv("course/Linear Regression/Weather.csv",low_memory=False)
```

```python
dataset.plot(x="Date",y="MaxTemp",style="o")
plt.title("Max Temp")
plt.xlabel("Day")
plt.ylabel("Maxtemp")
plt.show()
```

```python
dataset["MaxTemp"].replace(-17.77777778,method='pad',inplace=True)
```

```python
arange = np.arange(1,119041,1, dtype=int)
arange = arange.reshape(-1,1)
dataset["Day"] = arange
```

```python
x = dataset["Day"].values.reshape(-1,1)
y = dataset["MaxTemp"].values.reshape(-1,1)
```

```python
dataset.plot(x="Day",y="MaxTemp",style="o")
plt.title("Max Temp")
plt.xlabel("Day")
plt.ylabel("Maxtemp")
plt.show()
```

```python
x_train, x_test = train_test_split(x,random_state=0,test_size=0.2)
y_train, y_test = train_test_split(y,random_state=0,test_size=0.2)
```

```python
#training
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#testing
y_predict = model.predict(x_test)
```

```python
#compare
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict,color="red",linewidth="2")
plt.show()
```

```python
#compare
yt = y_test.flatten()
yp = y_predict.flatten()
df = pd.DataFrame( {'Actually':yt,'Predicted':yp} )
df.head(20).plot(kind="bar",figsize=(16,10))
plt.show()
```

```python
#metrics
print("MAE = ",metrics.mean_absolute_error(y_test,y_predict)) 
print("MSE = ",metrics.mean_squared_error(y_test,y_predict))
print("RMSE = ",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
print("Score = ",metrics.r2_score(y_test,y_predict))
```
